```
import pandas as pd

import numpy as np

import cv2

from sklearn.neighbors import KNeighborsClassifier
```

```
target = [1,1,2,2,3,3]
```

```
# i = j = 0
# path = "./data/img/{}_{}.bmp".format(i+1,j+1)
# digit = cv2.imread(path)
# digit = cv2.cvtColor(digit,code=cv2.COLOR_BGR2GRAY)
# #print(digit[:,:,0])
# # 将三维数据转会为二维
# digit = digit.reshape(204,-1)
# print(digit.shape)
```

```
X = []
for i in range(3):
    for j in range(2):
        path = "./data/img/{}_{}.bmp".format(i+1,j+1)
        digit = cv2.imread(path)
        digit = cv2.cvtColor(digit,code=cv2.COLOR_BGR2GRAY)
        X.append(digit.reshape(204,-1))
```

```
X = np.array(X)
X.shape
```

```
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,target,test_size=0.1)
```

```
print(X_train.shape)
X_train = X_train.reshape(5,-1)
print(X_train.shape)
```

```

knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,y_train)
```

```
path = "./data/img/test.bmp"
digit = cv2.imread(path)
digit = cv2.cvtColor(digit,code=cv2.COLOR_BGR2GRAY)
```

```
digit = digit.reshape(1,-1)
digit.shape
```

```
knn.predict_proba(digit)
```

```
knn.score(X_test,y_test)
```
