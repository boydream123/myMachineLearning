```
import matplotlib.pyplot as plt

from sklearn.neighbors import KNeighborsClassifier

import pandas as pd

import numpy as np

from sklearn import datasets
```

```
# X,y = datasets.load_iris(True)
# X.shape

# 第二种导入数据方法
iris = datasets.load_iris()
X = iris['data']
y = iris['target']
# 150代表样本的数量,训练数据
X.shape
```

```
# 降维，切片
X = X[:,:2]
X.shape
```

```
plt.scatter(X[:,0],X[:,1],c=y)
```

```
knn = KNeighborsClassifier(n_neighbors=5)
```

```
knn.fit(X,y)
```

```
# 获取测试数据
# 横坐标4~8;纵坐标2.0~4.5
# 背景点，取出来，meshrgrid

x1 = np.linspace(4,8,100)
y1 = np.linspace(2,4.5,80)

X1,Y1 = np.meshgrid(x1,y1)
display(X1.shape,Y1.shape)
```

```
# 原始点
plt.scatter(X1,Y1)
```

```
# # 方法一：先进行转化为一维数据，然后再进行合并
# X1 = X1.reshape(-1,1)
# Y1 = Y1.reshape(-1,1)
# X_test = np.concatenate([X1,Y1],axis=1)
# X_test

# np.c_进行数据合并，ravel进行平铺化，即展开为一维数据
X_test = np.c_[X1.ravel(),Y1.ravel()]
X_test
```

```
%%time
y_ = knn.predict(X_test)
```

```
plt.scatter(X_test[:,0],X_test[:,1],c=y_)
```

```
# 自由指定颜色
from matplotlib.colors import ListedColormap
lc = ListedColormap(['#FFAAAA','#AAFFAA','#AAAAFF'])
lc2 = ListedColormap(['#FF0000','#00FF00','#0000FF'])
plt.scatter(X_test[:,0],X_test[:,1],c=y_,cmap=lc)
plt.scatter(X[:,0],X[:,1],c=y,cmap = lc2)
```

```
# 绘制轮廓面
plt.contourf(X1,Y1,y_.reshape(80,100),cmap=lc)
plt.scatter(X[:,0],X[:,1],c=y,cmap = lc2)
```
